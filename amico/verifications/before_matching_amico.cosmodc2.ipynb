```python
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
inpath = "/sps/lsst/groups/clusters/cluster_comparison_project/before_matching/"
outpath = "/sps/lsst/users/namourou/web/desc/cluster_comparison_project/amico.cosmoDC2.fzb/magy/"
#outpath1 = outpath + "amico.cosmoDC2.fzb/magy/"
#outpath2 = outpath + "amico.DC2.fzb/magy/"
#outpathvs = outpath + "amico.cosmoDC2.fzb.magy_vs_amico.DC2.fzb.magy/"
```

```python
c1 = Table.read(inpath + "halos/cosmoDC2/v0/Catalog.fits")
c2 = Table.read(inpath + "amico/cosmoDC2.fzb/magy/Catalog.fits")
c2_mb = Table.read(inpath + "amico/cosmoDC2.fzb/magy/Catalog_members.fits")
```

## Redshift distributions

```python
z_bins = np.linspace(0,2.0,41)
plt.figure(figsize = [6,5])
plt.hist(c1["z_cl"], bins = z_bins, label = "cosmoDC2/v0", histtype = "step")
plt.hist(c2["z_cl"], bins = z_bins, label = "amico/cosmoDC2.fzb/magy", histtype = "step")
plt.xlabel("$z$")
plt.title("Redshift distributions")
plt.ylabel("Number of clusters")
plt.xlim([0,2.0])
plt.legend()
plt.savefig(outpath + "redshift_distributions.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize = [6,5])
plt.hist(c1["z_cl"], bins = z_bins, label = "cosmoDC2/v0", histtype = "step", density = True)
plt.hist(c2["z_cl"], bins = z_bins, label = "amico/cosmoDC2.fzb/magy", histtype = "step", density = True)
plt.xlabel("$z$")
plt.title("Normalized redshift distributions")
plt.ylabel("Normalized number of clusters")
plt.xlim([0,2.1])
plt.legend()
plt.savefig(outpath + "redshift_distributions_normalized.png", bbox_inches='tight')
plt.show()
```

### SNR distributions

```python
snr_bins = np.linspace(0,10,41)
plt.figure(figsize = [6,5])
#plt.hist(c1["snr_cl"], bins = snr_bins, label = "cosmoDC2/DC2.masked", histtype = "step")
plt.hist(c2["snr_cl"], bins = snr_bins, label = "amico/cosmoDC2.fzb/magy", histtype = "step")
plt.xlabel("SNR")
plt.title("SNR distributions")
plt.ylabel("Number of clusters")
plt.ylim([0,50000])
plt.xlim([3,14])
plt.legend()
plt.savefig(outpath + "snr_distributions.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize = [6,5])
#plt.hist(c1["snr_cl"], bins = snr_bins, label = "cosmoDC2", histtype = "step", density = True)
plt.hist(c2["snr_cl"], bins = snr_bins, label = "amico/cosmoDC2.fzb/magy", histtype = "step", density = True)
plt.xlabel("SNR")
plt.title("SNR distributions")
plt.ylabel("Number of clusters")
plt.xlim([3,14])
plt.legend()
plt.savefig(outpath + "snr_distributions_normalized.png", bbox_inches='tight')
plt.show()
```

### $\lambda^*$ versus z

```python
plt.figure(figsize = [9,7])
#plt.scatter(c1["snr_cl"], bins = snr_bins, label = "cosmoDC2", histtype = "step", density = True)
plt.scatter(c2["z_cl"], c2["mass"], alpha = .7, s=7, label = "amico/cosmoDC2.fzb/magy")
plt.xlabel("$z$", fontsize = 15)
plt.title("$\lambda^*$ versus $z$", fontsize = 17)
plt.ylabel("$\lambda^*$", fontsize = 15)
plt.ylim([0,250])
plt.xlim([0,2])
plt.legend()
plt.savefig(outpath + "lambdastar_vs_z_strange_features.png", bbox_inches='tight')
plt.show()
```
