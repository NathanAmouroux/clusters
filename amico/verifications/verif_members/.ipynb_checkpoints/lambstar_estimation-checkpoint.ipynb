## Mstar cut

```python
cosmo_values = pd.read_csv('/pbs/home/n/namourou/workspace/side_codes/clusters/amico/verifications/verif_members/model_SUB_i_greta_t5_param.txt', sep="  ")
cosmo_values = Table.from_pandas(cosmo_values)
```

```python
display(cosmo_values[0])
```

```python
mb['dist'] = 0.0
for i, id in enumerate(cl_cat['# ID']):
    if i%500==0:
        print(i)
    assoc_cl = cl_cat[cl_cat['# ID'] == id]
    assoc_mb = mb[mb['ASSOC_ID']==id]
    ra_cl = assoc_cl['Xphys']
    dec_cl = assoc_cl['Yphys']
    z_cl = assoc_cl['Zphys']
    ra_mb = assoc_mb['ra']
    dec_mb = assoc_mb['dec']
    z_mb = assoc_mb['redshift']
    ra_dist = ra_cl[0] - ra_mb
    dec_dist = dec_cl[0] - dec_mb
    dist =  np.sqrt((ra_dist)**2+(dec_dist)**2)
    mb['dist'][mb['ASSOC_ID']==id] = dist
```

```python
zeros1 = np.zeros(12)
c1_mb_c = Table(zeros1, names = ('GALID','FIELD_PROB', 'ASSOC_ID', 'ASSOC_PROB', 'ra', 'dec', 'redshift', 'mag_g', 'mag_r', 'mag_i', 'mag_z', 'mag_y'))
for i in range(len(cosmo_values['z'])-1):
    if i%50==0:
        print(i)
    z_min = cosmo_values['z'][i]
    z_max = cosmo_values['z'][i+1]
    mag_i_max = cosmo_values['mstar'][i+1] + 1.5 
    r200 = cosmo_values['r200[deg]'][i+1]
    cdt1 = mb[(mb['redshift']<=z_max)*(mb['redshift']>=z_min)*(mb['mag_i']<mag_i_max)*(mb['dist']<r200)]
    c1_mb_c = vstack([c1_mb_c, cdt1])
```

```python
cl_cat['lambstarest'] = 0.0
cl_cat['lambstarest_nomsk'] = 0.0
rich = []
for i, id in enumerate(cl_cat['# ID']):
    if i%1000==0:
        print(i)
    cl_cat['lambstarest_nomsk'][cl_cat['# ID'] == id] = sum(c1_mb_c[c1_mb_c['ASSOC_ID']==id]['ASSOC_PROB'])
    cl_cat['lambstarest'][cl_cat['# ID'] == id] = cl_cat['lambstarest_nomsk'][cl_cat['# ID']==id]/(1-cl_cat[cl_cat['# ID']==id]['MSKFRC'][0])
```

```python
np.mean((cl_cat["LAMBSTAR"]-cl_cat["lambstarest"])/cl_cat["LAMBSTAR"])
```

```python
plt.scatter(c1_mb_c['redshift'][1:], c1_mb_c['mag_i'][1:], s = .5, alpha = .9, label = "cut_members")
plt.plot(cosmo_values['z'], cosmo_values['mstar']+1.5, marker = '+', markersize = .5, alpha = 1, color= 'orange', label = "mstar+1.5")
plt.xlabel("z")
plt.ylabel("mag_i")
plt.legend()
```
