```python
skyMap = butler.get("deepCoadd_skyMap")
```

```python
ra = 60.25
dec = -36.20
point = geom.SpherePoint(ra*geom.degrees, dec*geom.degrees)
tp = skyMap.findTractPatchList([point])
```

```python
tractInfo = tp[0][0]
```

```python
tract = tp[0][0].getId()
patch = tp[0][1][0].getIndex()
print(tract, patch)
```

```python
wcs = tractInfo.getWcs()
patchInfo = tractInfo.getPatchInfo(patch)
outerBbox = patchInfo.getOuterBBox()
```

```python
print(outerBbox.beginX, outerBbox.endX)
```

```python
patchName = str(patch[0]) + ',' + str(patch[1])
print(tract, patchName)
dataId = {'tract':tract, 'patch':patchName}

bandpass_color_map = {'green':'r', 'red':'i', 'blue':'g'}

exposures = {}
for bandpass in bandpass_color_map.values():
    dataId['filter'] = bandpass
    exposures[bandpass] = butler.get('deepCoadd', dataId=dataId)

rgb_im = rgb.makeRGB(*(exposures[bandpass_color_map[color]].getMaskedImage().getImage()
                       for color in ('red', 'green', 'blue')), Q=8, minimum = -0.1, dataRange=1.5, 
                       saturatedPixelValue=100,
                       xSize=None, ySize=None)
```

```python
display.setDefaultBackend("matplotlib")

fig = plt.figure(figsize=(15, 15))
disp = rgb.displayRGB(rgb_im, show=False)
#xlim = plt.gca().get_xlim()
#ylim = plt.gca().get_ylim()
xyIm = np.linspace(0, 4200, 8)
xBB = np.linspace(outerBbox.beginX, outerBbox.endX, 8)
yBB = np.linspace(outerBbox.beginY, outerBbox.endY, 8)
xyBB = [geom.Point2D(xBB[i], yBB[i]) for i in range(len(xBB))] 
raDec = wcs.pixelToSky(xyBB)
xTicks = [f"{geom.radToDeg(raDec[i].getRa()):.2f}" for i in range(len(raDec))]
yTicks = [f"{geom.radToDeg(raDec[i].getDec()):.2f}" for i in range(len(raDec))]
raDec = wcs.pixelToSky(xyBB)
_ = plt.xticks(xyIm, xTicks)
_ = plt.yticks(xyIm, yTicks)
```
