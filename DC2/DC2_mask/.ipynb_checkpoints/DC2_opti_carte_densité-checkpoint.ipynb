# Density map

```python
#ra_min, ra_max = min(gal['ra']), max(gal['ra'])
#dec_min, dec_max = min(gal['dec']), max(gal['dec'])
ra_min, ra_max = 60,60.5
dec_min, dec_max = -36.4, -36

# Dfinir la taille des carrs (en degrs)
cote_carr = 0.01  # Vous pouvez ajuster cette valeur

# Crer une grille de carrs
ra_bins = np.arange(ra_min, ra_max + cote_carr, cote_carr)
dec_bins = np.arange(dec_min, dec_max + cote_carr, cote_carr)

# Crer des tableaux pour stocker les densits de galaxies
densits = np.zeros((len(ra_bins) - 1, len(dec_bins) - 1), dtype=float)
```

```python
len(ra_bins), len(dec_bins)
```

```python
# Parcourir chaque carr et calculer la densit de galaxies
for i in range(len(ra_bins) - 1):
    for j in range(len(dec_bins) - 1):
        ra_min, ra_max = ra_bins[i], ra_bins[i + 1]
        dec_min, dec_max = dec_bins[j], dec_bins[j + 1]

        # Slectionner les galaxies dans le carr
        in_square = (gal['ra'] >= ra_min) & (gal['ra'] < ra_max) & \
                    (gal['dec'] >= dec_min) & (gal['dec'] < dec_max)

        # Calculer la densit de galaxies dans le carr
        densit = np.sum(in_square) / (cote_carr * cote_carr)

        densits[i, j] = densit   
    if i%10==0:
        print(i)
```

```python
ra_grid, dec_grid = np.meshgrid(ra_bins, dec_bins)

# Crez une figure de densit
plt.figure(figsize=(10, 6))
plt.pcolormesh(ra_grid, dec_grid, densits.T, cmap='viridis')
plt.colorbar(label='Densit de galaxies')
plt.title('Densit de galaxies dans le plan ra, dec')
plt.xlabel('RA')
plt.ylabel('Dec')
plt.show()
```

```python
mu = np.mean(densits)
sigma = np.std(densits)
print(mu,sigma)
```

```python
cdt = (densits.T <= mu-3*sigma)
```

```python
ra_pt = []
for i in range(len(ra_bins)-1):
    ra_pt.append((ra_bins[i]+ra_bins[i+1])/2) 
```

```python
dec_pt = []
for i in range(len(dec_bins)-1):
    dec_pt.append((dec_bins[i]+dec_bins[i+1])/2) 
```

```python
ra_pt_grid, dec_pt_grid = np.meshgrid(ra_pt, dec_pt)
```

```python
indices_infrieurs__la_moyenne = np.where(densits.T < mu-2*sigma)

# Maintenant, extrayez les valeurs de RA et Dec correspondant aux indices infrieurs  la moyenne
ra_mask = ra_pt_grid[indices_infrieurs__la_moyenne]
dec_mask = dec_pt_grid[indices_infrieurs__la_moyenne]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.scatter(gal['ra'],gal['dec'], s=0.1, alpha = .5, color = 'red')
plt.scatter(ra_mask, dec_mask, s=5, color = 'white')
plt.xlabel("ra")
plt.ylabel("dec")
plt.title('galaxies in tract')
plt.xlim([60,60.5])
plt.ylim([-36.4,-36])
for i in range(len(ra_mask)):
    rectangle = plt.Rectangle((ra_mask[i]-0.5*cote_carr, dec_mask[i]-0.5*cote_carr), cote_carr, cote_carr, fill=True, color='black')
    ax.add_patch(rectangle)
plt.savefig('copy_hehe.png', bbox_inches='tight')

#plt.legend()
```
