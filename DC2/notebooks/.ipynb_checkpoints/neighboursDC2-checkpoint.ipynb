```python
import numpy as np
```

```python
grid = [[5074-i for i in range(13)], [4860-i for i in range(13)], [4648-i for i in range(13)], [4440-i for i in range(12)] + [9999], [4236-i for i in range(13)]
       , [4034-i for i in range(12)] + [9999], [3837-i for i in range(13)], [3643-i for i in range(13)], [3453-i for i in range(12)] + [9999], [3268-i for i in range(13)]
       , [3086-i for i in range(13)], [2908-i for i in range(12)] + [9999]]
```

```python
grid.append([9999 for i in range(13)])
```

```python
from itertools import product

size = 13

def neighbours(cell):
    for c in product(*(range(n-1, n+2) for n in cell)):
        if c != cell and all(0 <= n < size for n in c):
            yield c
```

```python
grid[0][1]
```

```python
grid
```

```python
l_o_ngbts = {"tract" : [], "list_of_neighbour_tiles" : []}
```

```python
for i in range(13):
    for j in range(13):
        curr_t_l = []
        ngbts = list(neighbours((i,j)))
        curr_t = grid[i][j]
        for element in ngbts:
            k, l = element
            if grid[k][l] != 9999:
                curr_t_l.append(grid[k][l])
        if curr_t != 9999:
            print("number of neightbours for ", str(curr_t), "is", len(curr_t_l), "they are", curr_t_l)
            l_o_ngbts["tract"].append(curr_t)
            l_o_ngbts["list_of_neighbour_tiles"].append(curr_t_l)
```

```python
l_o_ngbts
```

```python
from astropy.table import Table
```

```python
table = Table(l_o_ngbts)
```

```python
table
```

```python
for i, element in enumerate(table['list_of_neighbour_tiles']):
    table["list_of_neighbour_tiles"][i] = ",".join(str(x) for x in element)
```

```python
table['list_of_neighbour_tiles'] = table['list_of_neighbour_tiles'].astype(bytes)
```
